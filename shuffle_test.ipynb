```python
from csv import reader
import utils
import SplitStan
import numpy
import random
```

```python
# Load a CSV file
def load_csv(filename):
    file = open(filename, "r")
    lines = reader(file)
    datalist = list(lines)
    # 去掉index和id
    for row in datalist:
        row.pop(0)
        # row.pop(0)
    # 数据第一行为列名
    cols_name = datalist.pop(0)
    # convert string attributes to integers
    for i in range(len(datalist[0])):
        utils.str_column_to_float(datalist, i)
    folds = utils.cross_validation_split(datalist, 10)
    fold = folds[-1]
    folds.remove(fold)
    train_set = sum(folds, [])
    test_set = list()
    for row in fold:
        row_copy = list(row.copy())
        test_set.append(row_copy)
        row_copy[-1] = None
    actual = [row[-1] for row in fold] 
    return utils.DataSet(cols_name, train_set), utils.DataSet(cols_name, test_set), actual
```

```python
train_DS, test_DS, test_actual = load_csv("all_32.csv")
```

```python
train_DS_data_copy = train_DS.data.copy()
test_DS_data_copy = test_DS.data.copy()

pro_rec = utils.ProcedureRecorder(len(train_DS_data_copy))
tree = utils.build_tree(train_DS_data_copy, SplitStan.gini_index)

predictions = list()
for row in test_DS_data_copy:
    prediction = utils.predict(tree, row)
    predictions.append(prediction)
accuracy = utils.accuracy_metric(test_actual.copy(), predictions)
print("shuffle no col, result %s" %(accuracy))
```

```python
col_num = len(train_DS.cols_name)
accurs = list()
for i in range(col_num):
    test_DS_data_copy = test_DS.data.copy()
    train_DS_data_copy = train_DS.suffle_col(i)
    pro_rec = utils.ProcedureRecorder(len(train_DS_data_copy))
    tree = utils.build_tree(train_DS_data_copy, SplitStan.gini_index)
    # utils.name_cols(DS.cols_name, tree)
    predictions = list()
    for row in test_DS_data_copy:
        prediction = utils.predict(tree, row)
        predictions.append(prediction)
    accuracy = utils.accuracy_metric(test_actual.copy(), predictions)
    accurs.append(accuracy)
    print("shuffle col %s, result %s" %(i, accuracy))
```
